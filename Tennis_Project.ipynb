```python
import pandas as pd , numpy as np
import os
```

```python
def List_Tables_Os_Walk(MainDir="",TargetFile="",FileType = ".parquet",Ignore_name="",Duplicate=False):
    if Duplicate:
        Table_Dir = []
        for FolderName, Subfolders, Filenames in os.walk(MainDir):
            # print("The current folder is: " + FolderName)
            for Filename in Filenames:
                if Filename.endswith(FileType) and Filename.startswith(TargetFile) and Ignore_name not in Filename:
                    Table_Dir.append(os.path.join(FolderName, Filename))
                    # print(os.path.join(FolderName, Filename).split("\\")[-1])
        return Table_Dir
    else:
        Table_Dir = set()
        for FolderName, Subfolders, Filenames in os.walk(MainDir):
            # print("The current folder is: " + FolderName)
            for Filename in Filenames:
                if Filename.endswith(FileType) and Filename.startswith(TargetFile):
                    Table_Dir.add(os.path.join(FolderName, Filename))
                    # print(os.path.join(FolderName, Filename).split("\\")[-1])
        return list(Table_Dir)
```

```python
MainDir = r"D:\work\Daneshkar\DataScience\Project\Tennis Schema\Tennis Schema\tennis_data"
Table_Dir = List_Tables_Os_Walk(MainDir,TargetFile="pbp_",Ignore_name="",FileType = ".parquet",Duplicate=False)
df_all = pd.DataFrame()
print(len(Table_Dir))
i = 0
for Dir in Table_Dir:
    df = pd.read_parquet(Dir)
    date = Dir.split("\\")[-5]
    df['data_date'] = pd.to_datetime(date, format="%Y%m%d")
    print(df.columns)
    df_all = pd.concat([df_all, df], ignore_index=True)
    i += 1
    print(len(Table_Dir), i)
print(df_all.shape)
Csv_Path = r"D:\work\Daneshkar\DataScience\Project\Data_Csv\pbp_table_combined.csv"
df_all.to_csv(Csv_Path, index=False)
```
