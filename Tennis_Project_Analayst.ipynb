```python
#Question_1_V01
home_player_id_df = pd.read_csv(create_path_address("MatchHomeTeamInfo.csv"))["player_id"]
away_player_id_df = pd.read_csv(create_path_address("MatchAwayTeamInfo.csv"))["player_id"]

Total_Players = set(home_player_id_df) | set(away_player_id_df)
print(len(Total_Players))
```

```python
#Question_1_V02
home_player_id_df = pd.read_csv(create_path_address("MatchHomeTeamInfo.csv"))["player_id"]
away_player_id_df = pd.read_csv(create_path_address("MatchAwayTeamInfo.csv"))["player_id"]
Unique_Players = pd.concat([home_player_id_df, away_player_id_df], ignore_index=True).nunique()
print(Unique_Players)
```

```python
#Question_2_V01
home_away_df = pd.read_csv(create_path_address("MatchHomeTeamInfo.csv"))[["player_id","height","weight"]]
away_away_df = pd.read_csv(create_path_address("MatchAwayTeamInfo.csv"))[["player_id","height","weight"]]
merged = pd.concat([home_away_df, away_away_df], ignore_index=True).reset_index(drop=True).drop_duplicates()
player_weight_average = merged["weight"].mean()
player_height_average = merged["height"].mean()
print(f"The mean weight of players is : {player_weight_average}")
print(f"The mean height of players is : {player_height_average}")

```

```python
#Question_3_V01
Event_df = pd.read_csv(create_path_address("MatchEventinfo.csv"))[["match_id","winner_code"]]
home_df = pd.read_csv(create_path_address("MatchHomeTeamInfo.csv"))
away_df = pd.read_csv(create_path_address("MatchAwayTeamInfo.csv"))
merged_df = Event_df.merge(home_df[["match_id","player_id"]], on="match_id",how="left").merge(away_df[["match_id","player_id"]], on="match_id",how="left")
winner_map = {
    1: merged_df["player_id_x"],
    2: merged_df["player_id_y"]

}
merged_df["winner_player_id"] = merged_df["winner_code"].map(winner_map)
count_player_win = merged_df["winner_player_id"].value_counts()
top_player_id = count_player_win.idxmax()
home_df[home_df["player_id"] == top_player_id].iloc[0]
```

```python
#Question_3_V02
Event_df = pd.read_csv(create_path_address("MatchEventinfo.csv"))[["match_id","winner_code"]]
home_df = pd.read_csv(create_path_address("MatchHomeTeamInfo.csv"))
away_df = pd.read_csv(create_path_address("MatchAwayTeamInfo.csv"))
merged_df = Event_df.merge(home_df[["match_id","player_id"]], on="match_id",how="left").merge(away_df[["match_id","player_id"]], on="match_id",how="left")
merged_df["winner_player_id"] = np.where(
    merged_df["winner_code"] == 1,
    merged_df["player_id_x"],
    np.where(
        merged_df["winner_code"] == 2,
        merged_df["player_id_y"],
        np.nan
    )
)
count_player_win = merged_df["winner_player_id"].value_counts()
top_player_id = count_player_win.idxmax()
home_df[home_df["player_id"] == top_player_id].iloc[0]


```

```python
#Question_4_V01
Match_Time = pd.read_csv(create_path_address("MatchTimeInfo.csv"))
Match_Time["Total_Time"] = Match_Time[["period_1","period_2","period_3","period_4","period_5"]].fillna(0).sum(axis=1)
Longest_Match_id = Match_Time.loc[Match_Time["Total_Time"].idxmax()]["match_id"]
merged_df.loc[
    merged_df["match_id"] == Longest_Match_id,
    ["match_id", "player_id_x", "player_id_y", "winner_code"]
].iloc[0]

```
